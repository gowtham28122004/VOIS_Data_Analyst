```python
import pandas as pd,numpy as np,matplotlib.pyplot as plt,seaborn as sns
df = pd.read_csv(r'../VIOS/1719219914-Analysis of Super Store - DA (1) - 1719219914-Analysis of Super Store - DA (1).csv')
```

```python
# Result 1: Sales by Category
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(data=df, x='Category', y='Sales')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales ($)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()

# Calculate total sales by category
category_sales = df.groupby('Category')['Sales'].sum().round(2)
print("\
Total Sales by Category:")
print(category_sales)
```

```python
# Result 2: Average Profit by Region
plt.figure(figsize=(12, 6))
avg_profit_region = df.groupby('Region')['Profit'].mean()
avg_profit_region.plot(kind='bar')
plt.title('Average Profit by Region')
plt.xlabel('Region')
plt.ylabel('Average Profit ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

print("\
Average Profit by Region:")
print(avg_profit_region.round(2))
```

```python
# Result 3: Sales vs Profit by Category
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Sales', y='Profit', hue='Category')
plt.title('Sales vs Profit by Category')
plt.xlabel('Sales ($)')
plt.ylabel('Profit ($)')
plt.legend(title='Category')
plt.tight_layout()
plt.show()

# Calculate correlation between Sales and Profit
correlation = df['Sales'].corr(df['Profit']).round(3)
print("\
Correlation between Sales and Profit:", correlation)
```

### Romoval of Outliers

```python
sns.boxplot(y='Sales',data=df)
```

```python
def remove_outliers(col_names):
    sorted(col_names)
    Q1, Q3 = col_names.quantile([0.25,0.75])
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    uppre_bound = Q3 + 1.5 * IQR
    return lower_bound, uppre_bound
```

```python
min, max = remove_outliers(df['Sales'])
```

```python
df['Sales'] = np.where(df['Sales'] > max,max,df['Sales'])
```

```python
sns.boxplot(y='Sales',data=df)
```
